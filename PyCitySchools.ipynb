```python
# Add each prefix and suffix to remove to a list.
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]
```

```python
# Iterate through the words in the "prefixes_suffixes" list and replace them with an empty space, "".
for word in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word,"")
    
print(student_data_df.head(10))
```

```python
# Combine the both data/ DF into a single dataset. Merge() has to be on a shared column identical from each DF.
school_data_complete_df = pd.merge(student_data_df, school_data_df, on=["school_name", "school_name"])
school_data_complete_df.head()
```

```python
# Calculate the total budget.
total_budget = school_data_df["budget"].sum()
total_budget
```

```python
# Calculate the average reading score. In pandas we use the mean() method to calculate averages
average_reading_score = school_data_complete_df["reading_score"].mean()
average_reading_score
```

```python
# Calculate the average math score. In pandas we use the mean() method to calculate averages
average_math_score = school_data_complete_df["math_score"].mean()
average_math_score
```

```python
# Calculate the passing scores for the schools

passing_math = school_data_complete_df["math_score"] >= 70
passing_reading = school_data_complete_df["reading_score"] >= 70

passing_math.head(10)
```

```python
# To get the students who passed with >=70, filter through the school_data_complete_df for students who meet this passing requirement. ie"true"

# Get all the students who are passing math in the newly merged DataFrame "school_data_complete_df".
passing_math = school_data_complete_df[school_data_complete_df["math_score"] >= 70]
passing_math.head()
```

```python
# To get the students who passed with >=70, filter through the school_data_complete_df for students who meet this passing requirement. ie"true"

# Get all the students who are passing reading in the newly merged DataFrame "school_data_complete_df".
passing_reading = school_data_complete_df[school_data_complete_df["reading_score"] >= 70]
passing_reading.head()
```

```python
# Calculate the number of students passing math.
passing_math_count = passing_math["student_name"].count()

# Calculate the number of students passing reading.
passing_reading_count = passing_reading["student_name"].count()

passing_math_count, passing_reading_count
```

```python
# Calculate the percent that passed math.
passing_math_percentage = passing_math_count / float(student_count) * 100

# Calculate the percent that passed reading.
passing_reading_percentage = passing_reading_count / float(student_count) * 100

passing_math_percentage, passing_reading_percentage
```

```python
# Calculating the overall passing percentage, ie students who passed both math & reading
# Calculate the overall passing percentage.
overall_passing_percentage = (passing_math_percentage + passing_reading_percentage ) / 2

overall_passing_percentage
```

```python
# Creating a new DF by converting a list of dictionaries to a DF. Keys are column names and values are the metrics
# Adding a list of values with keys to create a new DataFrame.
district_summary_df = pd.DataFrame(
          [{"Total Schools": school_count,
          "Total Students": student_count,
          "Total Budget": total_budget,
          "Average Math Score": average_math_score,
          "Average Reading Score": average_reading_score,
          "% Passing Math": passing_math_percentage,
         "% Passing Reading": passing_reading_percentage,
        "% Overall Passing": overall_passing_percentage}])
district_summary_df
```

```python
# Chaining the map() &  format() functions. Basic syntax : df["column"] = df["column"].map("{:,}".format)
# Format the "Total Students" to have the comma for a thousands separator.

district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)

district_summary_df["Total Students"]
```

```python
# Format "Total Budget" to have the comma for a thousands separator, a decimal separator, and a "$".

district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)

district_summary_df["Total Budget"]
```

```python
# Format the columns.
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].astype(float).map("{:.1f}".format)

district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].astype(float).map("{:.1f}".format)

district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].astype(float).map("{:.0f}".format)

district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].astype(float).map("{:.0f}".format)

district_summary_df["% Overall Passing"] = district_summary_df["% Overall Passing"].astype(float).map("{:.0f}".format)


# Display the DF
district_summary_df
```

```python
# Reorder the columns in the order you want them to appear.
# This is also the way to filter out columns you don't need, and select columns in the order that you want.
new_column_order = ["Total Schools", "Total Students", "Total Budget","Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]

# Assign district summary df the new column order.
district_summary_df = district_summary_df[new_column_order]
district_summary_df
```
